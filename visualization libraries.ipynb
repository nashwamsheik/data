```python
import seaborn as sns
df=sns.load_dataset ("tips")
df.head()
```

```python
import matplotlib.pyplot as plt
plt.bar(df['day'],df['tip'])
plt.title("Bar Chart")

# Setting the X and Y labels
plt.xlabel ('Day')
plt.ylabel('Tip')
plt.show()
```

```python
plt.hist(df['total_bill'])
plt.title("Histogram")

plt.show()
```

```python
sns. lineplot(x="sex",y="total_bill",data=df)
plt.title( 'Title using Matplotlib Function')
```

```python
from scipy.stats import skew
data= np.array ([1, 2, 3, 4, 5, 6, 7, 100])
print("Skewness:", skew(data))
```

```python
from scipy.stats import kurtosis
data=np.array ([1, 2, 3, 4, 5,6, 7, 100])
print("Excess Kurtosis:", kurtosis (data))
print("Regular Kurtosis:", kurtosis(data, fisher=False))
```

```python
data = [2, 3, 4, 5, 6, 7, 8, 50]

# Plot simple histogram
plt.figure(figsize=(6, 4))
plt.hist(data, bins=8, color='lightcoral', edgecolor='black')
plt.title("Simple Kurtosis Example")
plt.xlabel("Value")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
