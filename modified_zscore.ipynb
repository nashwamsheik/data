```python
#python code to detect outliers using modified z score
```

```python
import pandas as pd
df=pd.read_csv("movie_revenues.csv")
df.tail(5)
```

```python
df.revenue.describe()
```

```python
#finding outliers using z score 
max_num = df.revenue.mean()+ 3*df.revenue.std()
min_num = df.revenue.mean()- 3*df.revenue.std()
print(max_num)
print(min_num)
```

```python
df2=df[(df.revenue>max_num) | (df.revenue<min_num)]
df2
```

```python
#now lets go to modified z score
```

```python
median1=df.revenue.median()
median1
```

```python
df['med']=abs(df.revenue-median1)
df.tail(5)
```

```python
MAD=df.med.median()
MAD
#median absolute deviation is MAD
```

```python
df['modified_zscore']=0.6745*(df.med/MAD)
df.tail(5)
```

```python
df2=df[df.modified_zscore > 3.5]
df2
```

```python
df2.revenue
```
